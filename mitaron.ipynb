```python
import pandas as pd
import datetime as dt
df = pd.read_csv("あつまれ どうぶつの森.csv").values
i = 0
dates = []
while True:
    try:
        dates.append(df[i][0])
    except:
        break
    i+=1
d=[]
for date in dates:
    a =date.split(".")
    date = dt.date(int(a[0]),int(a[1]),int(a[2]))
    d.append(date)
d.insert(0, dt.date(2017, 3, 2))
d
```

```python
s = []
c = 0
for item in d:
    dic={}
    for i in range(7):
        a = item + dt.timedelta(i)
        for b in t:
            if a == b:
                c+=1
    l=item + dt.timedelta(6)
    p = item.strftime('%Y.%m.%d') + "_" + l.strftime('%Y-%m-%d')
    dic["period"]= p
    dic["count_cum"]= c
    s.append(dic)
s
df = pd.DataFrame(s, columns=["period","count_cum"])
df.to_csv("softcount_2.csv")
```
