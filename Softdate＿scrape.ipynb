```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=options)
driver.implicitly_wait(10)
driver.get('https://www.nintendo.co.jp/software/switch/index.html?sftab=all&sfsort=adate')
a = driver.find_elements_by_class_name('nc3-c-softCard__name')
b = driver.find_elements_by_class_name('nc3-c-softCard__dateAndPrice')

dic = []
for i,item in enumerate(a):
  d = {}
  d["title"]=item.text
  d["date"]=b[i].text
  dic.append(d)
print(dic)

```

```python
import time
dic = []
for i in range(163):
  i+2
  driver.get(f'https://www.nintendo.co.jp/software/switch/index.html?sftab=all&sfsort=adate&spage={i}')
  a = driver.find_elements_by_class_name('nc3-c-softCard__name')
  b = driver.find_elements_by_class_name('nc3-c-softCard__dateAndPrice')
  for i,item in enumerate(a):
    d = {}
    d["title"]=item.text
    d["date"]=b[i].text
    dic.append(d)
  time.sleep(2)
print(dic)
```

```python
import pandas as pd
from google.colab import files

df = pd.DataFrame(dic, columns=["title","date"])
print(df)
df.to_csv("switch_softdates.csv")
files.download("switch_softdates.csv")
```

```python
data = []
for item in ml:
  try:
    driver.get(item["link"])
    e = driver.find_elements_by_id("entryBody")[0].text
    data.append(e.split("<p>")[-1])
    time.sleep(2)
  except:
    print(item["date"])
data
```

```python
a = []
for i,d in enumerate(data):
  p = {}
  g = []
  df = d.split("\n ")[1]
  f = df.split("\n")
  if f[0]=='':
    del f[0]
  for j,item in enumerate(f):
    try:
      j+=1
      b = {}
      b["title"]=item.split(f"{j} ",1)[1].split("週:")[0]
      b["week"]=item.split(f"{j} ",1)[1].split("週:")[1].split("計:")[0]
      b["total"]=item.split(f"{j} ",1)[1].split("週:")[1].split("計:")[1]
      g.append(b)
    except:
      print(item)
      g.append({})
  p["date"]=ml[i]["date"]
  p["data"]=g
  a.append(p)
```
